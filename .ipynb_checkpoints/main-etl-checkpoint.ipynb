```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
import pandas as pd
import numpy as np
```

```python
spark = SparkSession \
    .builder \
    .appName("how to read csv file") \
    .getOrCreate()
```

```python
def openDataFrame(data):
    df = spark.read.option("delimiter", ",") \
        .csv(f"data/{data}.csv", header=True)
    df.printSchema()
    return df
```
